```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: Pilot signals and received signal
# (You would replace these with actual data from the dataset)
N = 64  # Length of signal
pilot_symbols = np.array([1, -1, 1, -1, 1, -1, 1, -1])  # Example pilot symbols
received_signal = np.random.randn(N) + 1j * np.random.randn(N)  # Noisy received signal

# Assume the channel response is complex and has some delay
# Channel response (random for demonstration)
h = np.array([0.8 + 0.6j, 1 + 0.2j, 0.9 + 0.3j])

# Channel Estimation using Least Squares (LS)
# Assuming the received signal is a combination of pilot signals and channel response
# X (pilot matrix) and Y (received signal) are known
X = np.zeros((N, len(h)), dtype=complex)
Y = received_signal

# Create pilot matrix X (for simplicity, assume h is the channel response)
for i in range(len(h)):
    # Tile pilot_symbols to match the length of the received signal
    tiled_pilot_symbols = np.tile(pilot_symbols, N // len(pilot_symbols))
    X[:, i] = np.roll(tiled_pilot_symbols, i) # Now, tiled_pilot_symbols has shape (64,)

# Solve for the channel estimate using Least Squares
h_est = np.linalg.lstsq(X, Y, rcond=None)[0]

print("Estimated Channel Response: ", h_est)

# Plot the estimated vs actual channel
plt.figure(figsize=(8, 6))
plt.plot(np.real(h), label='Real part of actual h')
plt.plot(np.imag(h), label='Imaginary part of actual h')
plt.plot(np.real(h_est), '--', label='Real part of estimated h')
plt.plot(np.imag(h_est), '--', label='Imaginary part of estimated h')
plt.legend()
plt.title("Channel Estimation using Least Squares (LS)")
plt.show()
```
